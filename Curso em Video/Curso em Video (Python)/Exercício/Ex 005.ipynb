005-Faça um progama que leia um número inteiro e mostre na te la o seu Sucessor e seu Antecessor.

```python
n = int(input('Digite um número: '))
#Calculo
s = n + 1
a = n - 1
print(f'Analizando o valor {'\033[33m'}{n}{'\033[0m'} podemos perceber que seu Sucessor é {'\033[36m'}{s}{'\033[0m'} e o Antecessor {'\033[31m'}{a}{'\033[0m'}')
```

```python
#formula mais simples com o calculo dentro da chave de exibição.
n = int(input('Digite um número: '))
print(f'Analizando o valor {n} podemos perceber que seu Sucessor é {n + 1} e o Antecessor {n - 1}')
#Guanabara.
```
