008-Escreva um programa que leia um valor em metros e o exiba
convertido em centímetros e milimetros.

```python
m = float(input('Digite um valor em Metros:'))
c = m * 100
m = m * 1000
print(f'{m}M convertidos: \n{c} Centimetros. \n{m} Mlilímetros.')
```
