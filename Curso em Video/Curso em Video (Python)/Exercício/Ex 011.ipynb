011-Faça um programa que leia a largura e a altura de uma parede em metros, calcule a sua área e a quantidade de tinta necessária para pintá-la, sabendo que cada litro de tinta pinta uma área de 2 metros quadrados.

```python
lar = float(input('Digite a Largura da parede em metros: '))
alt = float(input('Digite a Altura da parede em metros: '))
area = lar * alt
ltinta = area / 2
print(f'''{'\033[1m'}Dimensão de ({'\033[31m'}{lar}{'\033[0m'} {'\033[36m'}x{'\033[0m'} {'\033[31m'}{alt}{'\033[0m'}) e sua área é de {'\033[36m'}{area}M²{'\033[0m'}.
Para pintar essa parede, você precisará de {'\033[36m'}{ltinta:.2f}L{'\033[0m'} de tinta.''')
```
