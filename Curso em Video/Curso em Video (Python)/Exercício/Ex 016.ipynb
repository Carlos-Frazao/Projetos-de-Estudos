016-Crie um programa que leia um número Real qualquer pelo teclado e mostre na tela a sua porção Inteira.

```python
from math import floor
n = float(input('Digite um número? '))
print(f'{'\033[4m'}O número{'\033[0m'} {'\033[31m'}{n}{'\033[0m'} {'\033[4m'}tem a parte inteira{'\033[0m'} {'\033[36m'}{floor(n)}{'\033[0m'}')
```

```python
from math import trunc
n = float(input('Digite um número: '))
print(f'O valor digitado foi {n} e sua porção inteira é {trunc(n)}')
```

```python
n = float(input('Digite um número: '))
print(f'O valor digitado foi {n} e sua porção inteira é {int(n)}')
```
