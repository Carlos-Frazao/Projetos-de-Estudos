019-Um professor quer sortear um dos seus quatro alunos para apagar o quadro. Fa√ßa um programa que ajude ele, lendo o nome dos alunos e escrevendo na tela o nome do escolhido.

```python
import random
print('=' * 46)
print('Digite o nome dos alunos para apagar o Quadro? ')
print('=' * 46)

n1 = str(input('Digite o nome do Aluno para paricipar: '))
n2 = str(input('Digite o nome do Aluno para paricipar: '))
n3 = str(input('Digite o nome do Aluno para paricipar: '))
n4 = str(input('Digite o nome do Aluno para paricipar: '))

lista = [n1, n2, n3, n4]
escolha = random.choice(lista)
print(f'O Aluno escolhido foi {'\033[36m'}{escolha}{'\033[0m'}.')
```
