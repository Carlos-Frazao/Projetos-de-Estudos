022-Crie um programa que leia o nome completo de uma pessoa e mostre: 
- O nome com todas as letras maiúsculas e minúsculas.
- Quantas letras ao todo (sem considerar espaços).
- Quantas letras tem o primeiro nome.

```python
nome = str(input('Digite seu nome completo? ')).strip()

print(f'''{'\033[1;33m'}Analisando seu nome...{'\033[0m'}
{'\033[35m'}Seu nome completo em letras Maiúscula,{'\033[0m'} {'\033[34m'}{nome.upper()}{'\033[0m'} 
{'\033[35m'}Seu nome completo em letras Minúcula,,{'\033[0m'} {'\033[34m'}{nome.lower()}{'\033[0m'}
{'\033[35m'}Quantas letras ao todo (sem considerar os espaços):{'\033[0m'} {'\033[31m'}{len(nome) - nome.count(' ')}{'\033[0m'}''')
#Quantas letras tem o primeiro nome: {nome.find(' ')}

separa = nome.split()
print(f'{'\033[35m'}Seu primeiro nome é{'\033[0m'} {'\033[34m'}{separa[0]}{'\033[0m'} {'\033[35m'}e tem{'\033[0m'} {'\033[31m'}{len(separa[0])}{'\033[0m'} {'\033[35m'}letras.')
```
