023-Faça um programa que leia um número de 0 a 9999 e mostre na tela cada um dos dígitos separados.

```python
num = int((input('Infome um número de 0 a 9999: ')))
unidade = num // 1 % 10
dezena = num // 10 % 10
centena = num // 100 % 10
milhar = num // 1000 % 10

print(f'''{'\033[1;33m'}Analisando o número: {num}{'\033[0m'} 
{'\033[1;36m'}Unidade: {unidade} 
Dezena: {dezena} 
Centena: {centena}
Milhar: {milhar}''')

#36 azul
```
