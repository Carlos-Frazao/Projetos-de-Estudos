027- Faça um programa que leia o nome completo de uma pessoa, mostrando em seguida o primeiro e o último nome separadamente.

```python
n = str(input('Digite seu nome completo? ')).strip()
nome = n.split()
p_nome = nome[0]
u_nome = nome[len(nome) -1 ]
print(f'''{'\033[1;33m'}Seu nome completo é{'\033[0m'} {'\033[1;4;36m'}{n}{'\033[0m'} 
{'\033[1;33m'}primeiro nome:{'\033[0m'} {'\033[1;4;36m'}{p_nome}{'\033[0m'}
{'\033[1;33m'}Último nome:{'\033[0m'} {'\033[1;4;36m'}{u_nome}{'\033[0m'}''')
```
