029-Escreva um programa que leia a velocidade de um carro. Se ele ultrapassar 80Km/h, mostre uma mensagem dizendo que ele foi multado. A multa vai custar R$7,00 por cada Km acima do limite.

```python
velo = float(input('Qual a velocidade do carro? '))
if velo <=80:
    print(f'{'\033[1;33m'}Tenha um bom dia! Dirija com segurança!{'\033[0m'}')
else:
    print(f'''{'\033[1;33m'}Você foi multado por exesso de velocidade, velocidade da Via é{'\033[0m'} {'\033[1;32m'}80Km/h.{'\033[0m'}
{'\033[1;33m'}Você deve pagar uma multa de{'\033[0m'} {'\033[1;31m'}{(velo - 80) * 7.00:.2f}R$!{'\033[0m'}
{'\033[1;33m'}Tenha um bom dia, Dirija com segurança!{'\033[0m'}''')
```
