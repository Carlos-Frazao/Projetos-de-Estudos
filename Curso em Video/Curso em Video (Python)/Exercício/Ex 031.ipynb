031-Desenvolva um programa que pergunte a distância de uma viagem em Km. Calcule o preço da passagem, cobrando R$0,50 por Km para viagens de até 200Km e R$0,45 parta viagens mais longas.


```python
distancia = float(input('Qual a distância da sua viagem? '))
preco = distancia * 0.50 if distancia <= 200 else distancia * 0.45
print(f'{'\033[4;33m'}O preço da sua passagem será de{'\033[0m'} {'\033[36m'}{preco:.2f}R${'\033[0m'}')
```

```python
km = float(input('Qual a distância da sua viajem: '))
if km <= 200:
    print(f'Sua viajem é de menos de 200Km, sendo assim o valor da passagem será de {km * 0.50:.2f}')
else:
    print(f'Como sua viagem é longa o preço da passagem é de {km * 0.45:.2f} com desconto')
    
```
