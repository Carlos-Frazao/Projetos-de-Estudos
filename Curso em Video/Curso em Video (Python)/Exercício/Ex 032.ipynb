032-Faça um programa que leia um ano qualquer e mostre se ele é bissexto.


```python
from datetime import date
a = int(input('Digite um ano qualquer? coloque 0 para analisar o ano atual: '))
if a == 0:
    a = date.today().year #pegando o ano atual da maquina
if a % 4 == 0 and a % 100 != 0 or a % 400 == 0:
    print(f'{'\033[1;33m'}O ano{'\033[0m'} {'\033[1;36m'}{a}{'\033[0m'} {'\033[1;33m'}é um ano Bissexto.{'\033[0m'}')
else:
    print(f'{'\033[1;33m'}O ano{'\033[0m'} {'\033[1;36m'}{a}{'\033[0m'} {'\033[1;33m'}não é Bissexto.{'\033[0m'}')
```
