035-Desenvolva um programa que leia o comprimento de três retas e diga ao usuário se elas podem ou não formar um triângulo.

```python
r1 = float(input('Digite o comprimento da primeira reta: '))
r2 = float(input('Digite o comprimento da segunda reta: '))
r3 = float(input('Digite o comprimeto da terceira reta: '))
print('-=' * 11)
print(f'{'\033[1;33m'}ANALISANDO O TRIÂNGULO{'\033[0m'}')
print('-=' * 11)
if r1 < r2 + r3 and r2 < r1 + r3 and r3 < r1 + r2:
    print(f'{'\033[1;33m'}As retas acima{'\033[0m'} {'\033[1;36m'}PODEM FORMAR {'\033[1;33m'}um trinângulo.{'\033[0m'}')
    if r1 == r2 == r3 and r1 == r3 == r2 and r2 == r1 == r3 and r2 == r3 == r1 and r3 == r1 == r2 and r3 == r2 == r1:
        print(f'{'\033[1;33m'}E Forma um Triângulo {'\033[1;36m'}Equilátero.{'\033[0m'}')
    elif r1 == r2 != r3 or r1 == r3 != r2 or r2 == r3 != r1:
        print(f'{'\033[1;33m'}E Forma um Triângulo {'\033[1;36m'}Isósceles.{'\033[0m'}')
    elif r1 != r2 != r3 or r2 != r1 != r3 or r3 != r1 != r2:
        print(f'{'\033[1;33m'}E Forma um Triângulo{'\033[0m'} {'\033[1;36m'}Escaleno.{'\033[0m'}')
else:
    print('As retas acima NÃO PODEM FORMAR triângulo.')
```
