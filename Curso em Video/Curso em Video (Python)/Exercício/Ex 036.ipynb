036-Escreva um programa para aprovar o empréstimo bancário para a compra de uma casa. Pergunte o valor da casa, o salário do comprador e em quantos anos ele vai pagar. A prestação mensal não pode exceder 30% do salário ou então o empréstimo será negado.


```python
print('=-' * 10)
print(f'{'\033[1;32m'}Emprestimo Bancário{'\033[0m'}')
print('=-' * 10)
casa = int(input('Qual o valor da casa: R$'))
salario = float(input('Seu salário: R$'))
quitacao = int(input('Quantos anos para quitar: '))
#calculo da prestação.
pres_mensal = casa / (quitacao * 12)
#calculo Execendecia de 30% do salario.
if pres_mensal < salario * 30 / 100:
    print(f'''{'\033[1;33m'}Para pagar uma casa de R${casa:.2f} em {quitacao} anos a prestação será de R${pres_mensal:.2f}.
Empréstimo{'\033[0m'} {'\033[1;32m'}Aprovado.{'\033[0m'}''')
else:
    print(f'''{'\033[1;33m'}Para pagar uma casa de R${casa:.2f} em {quitacao} anos a prestação ultrapassará 30% do seu salário.
Empréstimo{'\033[0m'} {'\033[31m'}Negado.{'\033[0m'}''')
```
