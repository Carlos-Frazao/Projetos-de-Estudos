038-Escreva um programa que leia dois números inteiros e compare-os. mostrando na tela uma mensagem:
mensagem:
- O primeiro valor é maior
- O segundo valor é maior
- Não existe valor maior, os dois são iguais

```python
n1 = int(input('Digite o Primeiro valor: '))
n2 = int(input('Digite o Segundo valor: '))
print(f'{'\033[7;30m'}-={'\033[0m'}' * 16)
print(f'{'\033[1;33m'}Comparando os Números Digitados.{'\033[0m'}')
print(f'{'\033[7;30m'}-={'\033[0m'}' * 16)
#Calculando o maior valor.
if n1 > n2:
    print(f'{'\033[1;33m'}O primeiro valor é{'\033[0m'} {'\033[1;36m'}Maior.{'\033[0m'}')
elif n1 < n2:
    print(f'{'\033[1;33m'}O Segundo valor é{'\033[0m'} {'\033[1;36m'}Maior{'\033[0m'}')
elif n1 == n2:
    print(f'{'\033[1;33m'}Não existe valor Maior, os dois são iguais.{'\033[0m'}')
```
