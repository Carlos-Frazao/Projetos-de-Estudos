042-Refaça o DESAFIO 035 dos triângulos, acrescentando o recurso de mostrar que tipo de triângulo será formado:
- EQUILÁTERO: todos os lados iguais
- ISÓSCELES: dois lados iguais, um diferente
- ESCALENO: todos os lados diferentes

```python
r1 = float(input('Digite o primeiro seguimento de reta: '))
r2 = float(input('Digite o segundo seguimento de reta: '))
r3 = float(input('Digite o terceiro seguimento de reta: '))
print('-=' * 11)
print(f'{'\033[1;4;33m'}ANALIZANDO O TRIÂNGULO{'\033[0m'}')
print('-=' * 11)
if r1 < r2 + r3 and r2 < r1 + r3 and r3 < r1 + r2:
    print(f'{'\033[1;32m'}As retas acima PODEM FORMAR um trinângulo.{'\033[0m'}')
    if r1 == r2 == r3 :
        print(f'{'\033[1;33m'}Forma um Triângulo{'\033[0m'} {'\033[1;36m'}Equilátero!{'\033[0m'}')
    elif r1 != r2 != r3 != r1:
        print(f'{'\033[1;33m'}Forma um Triângulo {'\033[1;36m'}Escaleno!{'\033[0m'}')
    else:
        print(f'{'\033[1;33m'}Forma um Triângulo{'\033[0m'} {'\033[1;36m'}Isósceles!{'\033[0m'}')
else:
    print(f'{'\033[1;31m'}As retas acima NÃO PODEM FORMAR um triângulo.{'\033[0m'}')
```
