043-Desenvolva uma lógica que leia o peso e a altura de uma pessoa, calcule seu Índice de Massa Corporal (IMC) e mostre seu status, de acordo com a tabela abaixo:
- IMC abaixo de 18,5: Abaixo do Peso
- Entre 18,5 e 25: Peso Ideal
- 25 até 30: Sobrepeso
- 30 até 40: Obesidade
- Acima de 40: Obesidade Mórbida

```python
peso = float(input('Qual seu Peso? (Kg) '))
altura = float(input('Qual é sua Altura? (m) '))
imc = peso / (altura ** 2) #Calculo de imc.
print(f'{'\033[1;33m'}O IMC dessa pessoa é de {imc:.1f}')
if imc < 18.5:
    print('Você está Abaixo do Peso!')
elif imc >= 18.5 and imc < 25:
    print(f'{'\033[1;36m'}PARABÉNS!{'\033[0m'} {'\033[0m'}Peso ideal.{'\033[0m'}')
elif imc >= 25 and imc < 30:
    print(f'{'\033[1;33m'}Você está em{'\033[0m'} {'\033[1;31m'}Sobrepeso.{'\033[0m'}')
elif imc >= 30 and imc < 40:
    print(f'{'\033[1;33m'}Você está em{'\033[0m'} {'\033[1;31m'}Obesidade!{'\033[0m'}')
elif imc >= 40:
    print(f'{'\033[1;31m'}Você está em Obesidade Mórbida! Cuidado Procure Tratamento.{'\033[0m'}')
```
