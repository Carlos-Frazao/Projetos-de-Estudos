045-Crie um programa que faça o computador jogar Jokenpô com você.

```python
from random import randint
from time import sleep
itens = ('Pedra', 'Papel', 'Tesoura')
computador = randint(0, 2)
print(f'''{'\033[1;33m'}Suas opções{'\033[0m'}
{'\033[1;36m'}[ 0 ] PEDRA 
[ 1 ] PAPEL
[ 2 ] TESOURA{'\033[0m'}''')
jogador = int(input('Qual é sua jogada? '))
print(f'{'\033[1;35m'}JO')
sleep(0.5)
print(f'{'\033[1;35m'}KEN{'\033[1;35m'}')
sleep(0.5)
print(f'{'\033[35m'}PO!!!{'\033[0m'}')
print('-=' * 11)
print(f'{'\033[1;33m'}Computador jogou {itens[computador]}{'\033[0m'}')
print(f'{'\033[1;33m'}Jogador jogou {itens[jogador]}{'\033[0m'}')
print('-=' * 11)
if computador == 0: #Computador jogou pedra.
    if jogador == 0:
        print(f'{'\033[1;37m'}EMPATE{'\033[0m'}')
    elif jogador == 1:
        print(f'{'\033[1;32m'}JOGADOR VENCE{'\033[0m'}')
    elif jogador == 2:
        print(f'{'\033[1;36m'}COMPUTADOR VENCE{'\033[0m'}')
    else:
        print(f'{'\033[1;31m'}Jogada Inválida!{'\033[0m'}')
elif computador == 1: #Computador jogou papel.
    if jogador == 0:
        print(f'{'\033[1;32m'}COMPUTADOR VENCE{'\033[0m'}')
    elif jogador == 1:
        print(f'{'\033[1;37m'}EMPATE{'\033[0m'}')
    elif jogador == 2:
        print(f'{'\033[1;32m'}JOGADOR VENCE{'\033[0m'}')
    else:
        print(f'{'\033[1;31m'}Jogada Inválida!{'\033[0m'}')
elif computador == 2: #Computador jogou tesoura.
    if jogador == 0:
        print(f'{'\033[1;32m'}JOGADOR VENCE{'\033[0m'}')
    elif jogador == 1:
        print(f'{'\033[1;36m'}COMPUTADOR VENCE{'\033[0m'}')
    elif jogador == 2:
        print(f'{'\033[1;37m'}EMPATE{'\033[0m'}')
#Guanabara.
```
