047-Crie um programa que mostre na tela todos os números pares que estão no intervalo entre 1 e 50.

```python
print(f'{'\033[1;33m'}Logo abaixo será exibido todos os números Pares de{'\033[0m'} {'\033[1;32m'}1{'\033[0m'} {'\033[1;33m'}até{'\033[0m'} {'\033[1;32m'}50.{'\033[m'}')
for c in range(1, 51):
    if c % 2 == 0:
        print(c)
```

```python
print('Logo abaixo será exibido todos os números Pares de 1 até 50.')
for c in range(2, 51, 2):
    print(c)
#Guanabara. Esse modo desgasta menos o processador porque ele faz apenas 25x de interação.
```
