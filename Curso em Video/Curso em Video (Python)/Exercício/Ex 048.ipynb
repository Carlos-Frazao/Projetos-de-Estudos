048-Faça um programa que calcule a soma entre todos os números que são múltiplos de três e que se encontram no intervalo de 1 até 500.

```python
s = 0
cont = 0
for c in range(1, 501, 2):
    if c % 3 == 0:
        cont = cont + 1
        s  = s + c
print(f'{'\033[1;33m'}A soma de todos os{'\033[0m'} {'\033[1;36m'}{cont}{'\033[0m'} {'\033[1;33m'}valores solicitados é{'\033[0m'} {'\033[1;36m'}{s}.{'\033[0m'}')
#Guanabara.
```
