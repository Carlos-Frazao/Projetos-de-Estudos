050-Desenvolva um programa que leia seis números inteiros e mostre a soma apenas daqueles que forem pares. Se o valor digitado for ímpar, desconsidere-o.

```python
cont = 0
s = 0
for c in range(1,7):
    n = int(input('Digite um número inteiro'))
    if n % 2 == 0:
        cont = cont + 1
        s = n + s
print(f'{'\033[1;36m'}Você informou{'\033[0m'} {'\033[32m'}{cont}{'\033[0m'} {'\033[1;33m'}números pares, A soma de todos os números Pares é{'\033[0m'} {'\033[32m'}{s}.{'\033[0m'}')
```
