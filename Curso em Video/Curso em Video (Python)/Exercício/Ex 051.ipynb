051-Desenvolva um programa que leia o primeiro termo e a razão de uma PA. No final, mostre os 10 primeiros termos dessa progressão.

```python
print('=' * 18)
print(f'{'\033[1;33m'}10 TEMOS DE UMA PA{'\033[0m'}')
print('=' * 18)
inicio = int(input('Primeiro termo: '))
razao = int(input('Progressão: '))
decimo = inicio + (10 - 1) * razao
for c in range(inicio, decimo + razao, razao):
    print(f'{'\033[1;32m'}{c}{'\033[0m'}')
print(f'{'\033[1;33m'}ACABOU{'\033[0m'}')
#Guanabara.
```
