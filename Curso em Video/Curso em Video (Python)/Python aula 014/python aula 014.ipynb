```python
for c in range(1, 11):
    print(c)
print('Fim!')
```

```python
c = 1 #Começa no 1
while c < 11: #Enquanto o contator for menor do que 10.
    print(c)
    c += 1 #Aqui ele calcula c mais 1, vai ficar repetindo até chegar em 10.
print('Fim!')
```

```python
n = 1
par = impar = 0
while n != 0:
    n  = int(input('Digite um valor: '))
    if n != 0: # se o (n) for diferente de (0) ele irá fazer os comandos abaixo, se for igual, irá terminar o programa.
        if n % 2 == 0:
            par += 1
        else:
            impar += 1
print(f'Você digitou {par} números pares e {impar} números ímpares.')
```
