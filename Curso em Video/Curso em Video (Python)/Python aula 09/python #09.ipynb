```python
frase = 'Curso em Vídeo Python'

frase.count('o')
#.count() - Significa contador.
#frase.count(str('o')) Outra maneira especificada

frase.count('o', 0, 13)
#Contagem já com o fatiamento
```

```python
frase = 'Curso em Vídeo Python'
frase.replace('Python', 'Android')
#.replace() Subistituiçã o.

frase.upper()
#Deixaa tudo Maíuscula.

frase.lower()
#Deixa tu em Minúscula

frase.capitalize()
#Deixa só a primeira letra da frase em maiúscula.

frase.title()
#Na frase onde tiver espaço, ele coloca a proxima frase em maiúscula.

frase.strip()
#Remove espaços exedentes.

frase.rstrip()
#Rmove espaços do lado direito.

frase.lstrip()
#Remove espaços do lado esquerdo.

frase.split()
#Onde tiver espaços ele cria uma divisão. E outra Divisão.

'-'.join(frase)
#Subistituia os espaços pelo '-' informado.

frase.rfind()
#começar a procurar do lado direito.
```
