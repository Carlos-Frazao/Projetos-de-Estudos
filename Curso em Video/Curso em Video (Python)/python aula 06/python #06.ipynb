### Tipos Primitivos

```python
inteiro = (int) = 7, -4, 0, 3, 9
float = (float) = 3.14, -0.001, 2.0, 0
string = (str) = "hello", "world", "", "123"
bool = (bool) = True, False
list = (list) = [1, 2, 3], [], ["a", "b", "c"]
```

### Tipos

```python
print("Demostração de tipos primitivos em python:")
n = 9
print(type(n))
n = 3.14
print(type(n))
n = "Hello, World!"
print(type(n))
n = True
print(f"{type(n)}\n")

print("Verificando tipos com funções específicas:")
algo = "3"
print(algo.isnumeric()) # Verifica se é numérico
algo = str("Oi")
print(algo.isalpha()) # Verifica se é alfabético
algo = "Hello123"
print(algo.isalpha()) # Verifica se é alfabético
algo = "1234"
print(algo.isalnum()) # Verifica se é alfanumérico
algo = "Oi"
print(algo.isupper()) # Verifica se é maiúsculo
algo = "oi"
print(algo.islower()) # Verifica se é minúsculo
```
