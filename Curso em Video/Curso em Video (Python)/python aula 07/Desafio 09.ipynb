```python
def main():
    try:
        # Entrada de dados 
        n = int(input("Informe um número, será mostrado a tabuada: "))

        cont = 0
        while(cont < 11):
            print(f"{n} X {cont} = {cont*n}")
            cont+=1

    except ValueError:
        print("Informe um valor inteiro!")

if __name__ == "__main__":
    main()
```
