```python
def main():
    # Entrada de dados
    real = float(input("Quanto dinheiro você tem: R$"))
    cotacao = float(input("Qual a cotação do dolar: "))

    # Saída de dados com processamento
    print(f"Com R${real} você consegue comprar ${real/cotacao:.2f} dolares!")
if __name__ == "__main__":
    main()
```
