```python
def main():
    salario = float(input("Informe o salário do funcionário: R$"))
    # Saída de dados com o aumento.
    print(f"Um funcionário que ganha R${salario:.2f}, com 15% de aumento, passa a ganhar R${salario+(salario*15/100):.2f}")
if __name__ == "__main__":
    main()
```

```python
def aumento_salario(salario_atual, porcentagem_aumento):
    return salario_atual + (salario_atual * porcentagem_aumento / 100)

def main():
    try:
        salario = float(input("Informe o salário do funcionário: R$"))
        porcentagem = float(input("Informe a porcentagem de aumento: "))
        # Saída de dados com o aumento.
        print(f"Um funcionário que ganha R${salario:.2f}, com {porcentagem}% de aumento, passa a ganhar R${aumento_salario(salario, porcentagem):.2f}")
    except ValueError:
        print("Informe o valor monetário com . e a porcentagem sem o simbolo")
if __name__ == "__main__":
    main()
```
