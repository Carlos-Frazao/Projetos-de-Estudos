### Operadores Aritméticos

* **`+`** : Adição
* **`-`** :  Subtração
* **`\*`** : Multiplicação
* **`/`** : Divisão
* `**` : Potência
* **`//`** : Divisão inteira
* **`%`** : Resto da divisão

```python
def main():
    print(f"Adição: 5 + 2 = {5 + 2}")
    print(f"Subtração: 5 - 2 = {5 - 2}")
    print(f"Multiplicação: 5 * 2 = {5 * 2}")
    print(f"Divisão: 5 / 2 = {5 / 2}")
    print(f"Potência: 5 ** 2 = {5 ** 2}")
    print(f"Divisão inteira: 5 // 2 = {5 // 2}")
    print(f"Módulo resto da divisão: 5 % 2 = {5 % 2}")
if __name__ == "__main__":
    main()
```

### Ordem de Precedência

1. **`()`** - Parênteses
2. **`**`** - Potência (Exponenciação)
3. **`*`**, **`/`**, **`//`**, **`%`** - Multiplicação, Divisão, Divisão Inteira e Resto
4. **`+`**, **`-`** - Adição e Subtração


```python
def main():
    print("Exemplos de Operadores Aritméticos em Python:")
    print(f"5 + 3 * 2 = {5 + 3 * 2}")
    print(f"3 * 5 + 4 ** 2 = {3 * 5 + 4 ** 2}")
    print(f"3 * (5 + 4) ** 2 = {3 * (5 + 4) ** 2}")

if __name__ == "__main__":
    main()
```
