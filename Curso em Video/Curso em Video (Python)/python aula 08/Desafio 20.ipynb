```python
from random import shuffle

def main():
    c = 1
    alunos = [] # Lista para armazenar os nomes dos alunos
    while c < 5:
        alunos.append(input(f"Informe o nome do aluno {c}: "))
        c += 1
    
    print("A nova ordem de apresentação é:")

    random.shuffle(alunos) # Embaralha a lista de alunos

    # O enumerate começa a contar do 1 e percorre a lista automaticamente
    for i, aluno in enumerate(alunos, start=1):
        print(f"{i} - {aluno}")

if __name__ == "__main__":
    main()
```
