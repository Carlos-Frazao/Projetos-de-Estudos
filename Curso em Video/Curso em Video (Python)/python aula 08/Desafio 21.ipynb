```python
import pygame 

def main(): 
    pygame.init() # Inicia o módulo pygame

    try:
        pygame.mixer.music.load("Ex021.MP3") # Carrega o arquivo de música
        pygame.mixer.music.play() # Reproduz a música

        print("Pressione Enter para parar a música...")
        input()  # Aguarda o usuário pressionar Enter

    except pygame.error:
        print("Ocorreu um erro ao inicializar o pygame.")


if __name__ == "__main__":
    main()
```
