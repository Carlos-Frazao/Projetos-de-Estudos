```python
import math
n = int(input('Digite um número: '))
raiz = math.sqrt(n)
print(f'A raiz de {n} é igual a {raiz:.2f}')
```

```python
from math import sqrt, floor
n = int(input('Digite um número: '))
raiz = sqrt(n)
print(f'A raiz de {n} é igual a {floor(raiz)}')
```
