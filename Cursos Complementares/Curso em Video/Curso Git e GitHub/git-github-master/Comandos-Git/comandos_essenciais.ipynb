### Configuração e Início  

* **git config --global user.name "Nome"**: Identifica quem está fazendo as alterações.

* **git config --global user.email "email@exemplo.com"**: Vincula os commits à sua conta do GitHub.

* **git init**: Inicia um repositório Git local em uma pasta.

* **git clone [URL]**: Copia um repositório remoto existente para o seu computador.  

### O Fluxo Diário (Modificar e Salvar)  

* **git status**: Mostra o estado atual dos arquivos (quais foram alterados ou não salvos).

* **git add [arquivo]**: Prepara um arquivo específico para ser salvo.

* **git add .**: Prepara todos os arquivos alterados de uma vez.

* **git commit -m "mensagem"**: Salva permanentemente a "foto" do seu código com uma  mensagem explicativa sobre o que foi alterado.

### Sincronização com o GitHub  

* **git remote add origin [URL]**: Conecta seu Git local ao repositório no GitHub.

* **git push origin [branch]**: Envia seus commits locais para o servidor online.

* **git pull**: Puxa as novidades do servidor e atualiza seu código local.

* **git fetch**: Baixa as atualizações do servidor, mas não as aplica automaticamente (mais seguro).  

### Ramificações (Branches) - Para não "atropelar" o código 

* **git branch**: Lista todas as branches do seu projeto.

* **git checkout -b [nome]**: Cria uma nova branch e muda para ela instantaneamente.

* **git checkout [nome]**: Alterna entre branches existentes.

* **git merge [nome]**: Une as alterações de uma branch específica à que você está usando  

### Histórico e Correções  

* **git log**: Exibe o histórico de todos os commits feitos no projeto.

* **git diff**: Mostra detalhadamente o que mudou no código antes de você salvar.

* **git reset --hard [HASH]**: Volta o projeto para um estado anterior específico, descartando tudo o que veio depois.

* **git restore [arquivo]**: Desfaz as alterações feitas em um arquivo que ainda não foi salvo (commitado).
