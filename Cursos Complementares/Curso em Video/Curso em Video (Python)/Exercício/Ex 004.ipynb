004-Faça um programa que leia algo pelo teclado e mostre na tela o seu tipo primitivo
e todas as informações possíveis sobre ele.


```python
a = (input('Digite algo: '))
print(f'Logo a baixo estará todos os tipos primitivos de')
print(a)

t0 = a.isalnum()
t1 = a.isalpha()
t2 = a.islower()
t3 = a.isnumeric()
t4 = a.isspace()
t5 = a.istitle()
t6 = a.isupper()


print(f'É alfanúmerico: {t0}')
print(f'Contém apenas letras: {t1}')
print(f'Está em letra minúscula: {t2}')
print(f'É númerico: {t3}')
print(f'Tem espaços {t4}')
print(f'A primeira letra é Maiúscula? {t5}')
print(f'Está tudo em Maiúscula {t6}')
```

```python
a = input('Digite algo: ')
print(f'''{'\033[1;33m'}Logo abaixo estará todos os tipos primitivo de...{'\033[0m'}
{'\033[4;31m'}{a}{'\033[0m'}''')
cores = {'limpa' : '\033[0m', 
         'azul' : '\033[34m',
         'verde' : '\033[32m'}
print(f'{cores['azul']}O tipo primitivo desse valor é?{cores['limpa']}{cores["verde"]}',type(a))
print(f'{cores["azul"]}Só tem espaços?{cores["limpa"]}{cores["verde"]}', a.isspace())
print(f'{cores["azul"]}É um número?{cores["limpa"]}{cores["verde"]}', a.isnumeric())
print(f'{cores["azul"]}É alfabético?{cores["limpa"]}{cores["verde"]}', a.isalpha())
print(f'{cores["azul"]}É alfanumérico?{cores["limpa"]}{cores["verde"]}', a.isalnum())
print(f'{cores["azul"]}Está em maiúscula?{cores["limpa"]}{cores["verde"]}', a.isupper())
print(f'{cores["azul"]}Está em maiúscula?{cores["limpa"]}{cores["verde"]}', a.islower())
print(f'{cores["azul"]}Está capitalizada?{cores["limpa"]}{cores["verde"]}', a.istitle())
#Guanabara.
```
