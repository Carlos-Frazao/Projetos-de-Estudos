006-Crie um algoritmo que leia um número e mostre o seu dobro, triplo e raiz quadrada.

```python
n = int(input('Digite um número: '))
print(f'O dobro de {n} vale {n * 2}. \nO triplo de {n} vale {n * 3 }. \nA raiz quadrada de {n} vale {n ** (1/2):.2f}.')
```

```python
import math
n = float(input('Digite um número: '))
d = n * 2
t = n * 3
r = math.pow(n, 1/2)
print(f'{'\033[4;34;40m'}O Dobro de {n} vale {d}. ')
print(f'O Triplo de {n} vale {t}. ')
print(f'A Raiz Quadrada de {n} vale {r:.2f}. ')
#Guanabra.
```
