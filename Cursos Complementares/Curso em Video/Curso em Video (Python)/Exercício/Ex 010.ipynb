010-Crie um programa que leia quanto dinheiro uma pessoa tem na carteira e mostre quantos dólares ela pode comprar.

```python
real = float(input('Quanto dinheiro você tem? R$'))
#conv = c / 5.35
conv = real / 3.27 #conversão
print(f'{'\033[4m'}{'\033[1m'}Com {real:.2f}R$ você consegue comprar US${conv:.2f} Dolares.{'\033[0m'}')
```
