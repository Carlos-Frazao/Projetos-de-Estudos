013-Faça um algoritmo que leia o salário de um funcionário e mostre seu novo salário, com 15% de aumento.

```python
salario = float(input('Digite o valor do seu salário? R$'))
novo_salario = salario + (salario * 0.15)
print(f'{'\033[1m'}Seu salário de {'\033[31m'}{salario:.2f}{'\033[0m'}R$ com {'\033[36m'}15%{'\033[0m'} de Aumento, Fica no Valor de {'\033[32m'}{novo_salario:.2f}{'\033[0m'}R$.')
```
