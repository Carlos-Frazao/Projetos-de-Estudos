014-Escreva um programa que converta uma temperatura digitando em graus Celsius e converta para graus Fahrenheit.

```python
c = float(input('Informe a temperatura em °C: '))
conv = (c * 9/5) + 32
print(f'A temperatura de {c}°C corresponde a {conv}°F!')
```

```python
c = float(input('Informe a temperatura em °C: '))
conv = (c * 9/5) + 32
if c < 30:
    print(f'{'\033[1m'}A temperatura de {'\033[36m'}{c}°C{'\033[0m'} {'\033[1m'}corresponde a {'\033[33m'}{conv}°F{'\033[0m'}!')
else:
    print(f'{'\033[1m'}A temperatura de {'\033[31m'}{c}°C{'\033[0m'} {'\033[1m'}corresponde a {'\033[33m'}{conv}°F{'\033[0m'}!')
```
