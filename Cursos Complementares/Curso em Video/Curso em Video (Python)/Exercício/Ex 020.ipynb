020-O mesmo professor do desafio 019 quer sortear a ordem de apresentação de trabalhos dos alunos. 
Faça um programa que leia o nome dos quatro alunos e mostre a ordem sorteada.

```python
import random
print('=' * 33)
print('Ordem de apresentação de Trabalho')
print('=' * 33)

n1 = str(input('Digite o nome do Aluno: '))
n2 = str(input('Digite o nome do Aluno: '))
n3 = str(input('Digite o nome do Aluno: '))
n4 = str(input('Digite o nome do Aluno: '))

lista = [n1, n2, n3, n4]
random.shuffle(lista)
print(f'A ordem de apresentacão  \n{lista}')
```

```python
from random import shuffle
print('=' * 33)
print(f'{'\033[4m'}Ordem de apresentação de Trabalho{'\033[0m'}.')
print('=' * 33)

n1 = str(input('Digite o nome do Aluno: '))
n2 = str(input('Digite o nome do Aluno: '))
n3 = str(input('Digite o nome do Aluno: '))
n4 = str(input('Digite o nome do Aluno: '))

lista = [n1, n2, n3, n4]
shuffle(lista)
print(f'{'\033[4m'}A ordem de apresentacão será{'\033[0m'}.\n{'\033[36m'}{lista}{'\033[0m'}')
```
