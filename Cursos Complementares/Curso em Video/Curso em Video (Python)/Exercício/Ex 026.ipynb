026-Faça um programa que leia uma frase pelo teclado e mostre quantas vezes aparece a letra "A", em que posição ela aparece a primeira vez e em que posição ela aparece a última vez.

```python
frase = str(input('Digite uma frase: ')).upper().strip()
print(f'''{'\033[1;33m'}Quantas vezes a letra {'\033[4m'}A{'\033[0m'} {'\033[1;33m'}aparece na frase?{'\033[0m'} {'\033[31m'}{frase.count('A')}{'\033[0m'} {'\033[1;33m'}vezes{'\033[0m'}.
{'\033[1;33m'}A primeira letra {'\033[4m'}A{'\033[0m'} {'\033[1;33m'}apareceu na posição:{'\033[0m'} {'\033[31m'}{frase.find('A') + 1}{'\033[0m'}
{'\033[1;33m'}A última letra {'\033[4m'}A{'\033[0m'} {'\033[1;33m'}apareceu na posição:{'\033[0m'} {'\033[31m'}{frase.rfind('A') + 1}{'\033[0m'}''')
```
