030-Crie um programa que leia um número inteiro e mostre na tela se ele é PAR ou ÍMPAR.

```python
n = int(input('Digite um número inteiro: '))
#par = n % 2 == 0
if n % 2 == 0:
    print(f'{'\033[1;33m'}O número{'\033[0m'} {'\033[1;4;36m'}{n}{'\033[0m'} {'\033[1;33m'}é{'\033[0m'} {'\033[1;36m'}par.{'\033[0m'}')
else:
    print(f'{'\033[1;33m'}O número{'\033[0m'} {'\033[1;4;34m'}{n}{'\033[0m'} {'\033[1;33m'}é{'\033[0m'} {'\033[1;34m'}ímpar.{'\033[0m'}')
```
