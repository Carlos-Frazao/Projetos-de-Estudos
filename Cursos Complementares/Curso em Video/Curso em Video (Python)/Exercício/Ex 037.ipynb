037-Escreva um programa em Python que leia um número inteiro qualquer e peça para o usuário escolher qual será a base de conversão: 1 para binário, 2 para octal e 3 para hexadecimal.

```python
n = int(input('Digite um número inteiro qualquer: '))
print('-=' * 15)
print(f'{'\033[1;33m'}Escolha uma das bases para conversão.{'\033[0m'}')
print('-=' * 15)
print(f'''{'\033[1;32m'}[ 1 ] Conversão para Binário.
[ 2 ] Conversão para Octal.
[ 3 ] Conversão para Hexadecimal.{'\033[0m'}''')
opcao = int(input('Sua opção: '))
if opcao == 1:
    print(f'{'\033[1;36m'}{n} Convertido para Binário é igual a {bin(n)[2:]}{'\033[0m'}') #bin() Binário.
elif opcao == 2:
    print(f'{'\033[1;36m'}{n} Convertido para Octal é igual a {oct(n)[2:]}{'\033[0m'}') #oct() Octal.
elif opcao == 3:
    print(f'{'\033[1;36m'}{n} Convertido para Hexadecimal é igual a {hex(n)[2:]}{'\033[0m'}') #hex() Hexadecimal.
else:
    print(f'{'\033[1;31m'}Opção invalida. Tente novamente{'\033[0m'}')
```
