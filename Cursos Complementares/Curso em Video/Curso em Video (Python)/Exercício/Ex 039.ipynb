039-Faça um programa que leia o ano de nascimento de um jovem e informe, de acordo com a sua idade, se ele ainda vai se alistar ao serviço militar, se é a hora exata de se alistar ou se já passou do tempo do alistamento. Seu programa também deverá mostrar o tempo que falta ou que passou do prazo.

```python
from datetime import date
ano_nascimento = int(input('Jovem, qual seu ano de nascimento: '))
#Alistamento com >=18.
ano_atual = date.today().year
sexo = str(input('Qual seu sexo? '))
print(f'{'\033[1;33m'}Quem nasceu em {ano_nascimento} tem {ano_atual - ano_nascimento} anos.{'\033[0m'}')
if 'masculino' == sexo.lower():
    if (ano_atual - ano_nascimento) >= 19:
        print(f'{'\033[1;33m'}Seu alistamento foi em{'\033[0m'} {'\033[1;31m'}{ano_nascimento + 18}{'\033[0m'}')
        print(f'{'\033[1;33m'}passou{'\033[0m'} {'\033[1;31m'}{(ano_atual - ano_nascimento) - 18}{'\033[0m'} {'\033[1;33m'}anos do seu alistamento meu nobre.{'\033[0m'}')
    elif (ano_atual - ano_nascimento) == 18:
        print(f'{'\033[1;33m'}Hora de se alistar meu nobre{'\033[0m'}')
    elif (ano_atual - ano_nascimento) < 18:
        print(f'{'\033[1;33m'}Seu alistamento será em{'\033[0m'} {'\033[1;32m'}{ano_nascimento + 18}{'\033[0m'}')
        print(f'{'\033[1;33m'}Falta{'\033[0m'} {'\033[1;32m'}{18 - (ano_atual - ano_nascimento)}{'\033[0m'} {'\033[1;33m'}anos para você se alistar meu caro.{'\033[0m'}')
else:
    print(f'{'\033[1;33m'}Você não precisa fazer Alistamento Militar.{'\033[0m'}')
```
