040-Crie um programa que leia duas notas de um aluno e calcule sua média, mostrando uma mensagem no final, de acordo com a média atingida:
- Média abaixo de 5.0: REPROVADO
- Média entre 5.0 e 6.9: RECUPERAÇÃO
- Média 7.0 ou superior: APROVADO

```python
n1 = float(input('Digite a primeira nota do Aluno: '))
n2 = float(input('Digite a segunda nota do Aluno:'))
m = (n1 + n2) / 2
print(f'{'\033[4m'}Média do aluno: {m:.1f}{'\033[0m'}')
if m < 5.0:
    print(f'{'\033[1;31m'}Reprovado!{'\033[0m'}')
elif m > 5.0 and m < 6.9:
    print(f'{'\033[1;33m'}Recuperação!{'\033[0m'}')
elif m >= 7.0:
    print(f'{'\033[1;36m'}Aprovado!{'\033[0m'}')
```
