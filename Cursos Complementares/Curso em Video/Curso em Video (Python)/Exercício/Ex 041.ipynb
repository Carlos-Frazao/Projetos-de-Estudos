041-A Confederação Nacional de Natação precisa de um programa que leia o ano de nascimento de um atleta e mostre sua categoria, de acordo com a idade:
- Até 9 anos: MIRIM
- Até 14 anos: INFANTIL
- Até 19 anos: JÚNIOR
- Até 25 anos: SÊNIOR
- Acima de 25 anos: MASTER

```python
nascimento = int(input('Digite o ano de nascimento do atleta? '))
print(f'{'\033[4m'}Sua categoria é{'\033[0m'}')
ano = 2024 - nascimento
if ano <= 9:
    print(f'{'\033[1;36m'}Atleta mirim.{'\033[0m'}')
elif ano >= 10 and ano <= 14:
    print(f'{'\033[1;36m'}Atleta infantil.{'\033[0m'}')
elif ano >= 15 and ano <= 19:
    print(f'{'\033[1;36m'}Atleta junior.{'\033[0m'}')
elif ano <= 25:
    print(f'{'\033[1;36m'}Atleta sênior.{'\033[0m'}')
elif ano > 25:
    print(f'{'\033[1;36m'}Atleta master.{'\0333[0m'}')
```
