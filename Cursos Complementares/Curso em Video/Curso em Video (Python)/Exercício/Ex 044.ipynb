044-Elabore um programa que calcule o valor a ser pago por um produto, considerando o seu preço normal e condição de pagamento:
- à vista dinheiro/cheque: 10% de desconto
- à vista no cartão: 5% de desconto
- em até 2x no cartão: preço formal
- 3x ou mais no cartão: 20% de juros

```python
print(f'''{'\033[1;33m'}Condições de Pagamento.{'\033[0m'}
{'\033[1;36m'}(1) À Vista no Dinheiro / Cheque: 10% de Desconto.
(2) À Vista no Cartão: 5% de Desconto.
(3) Em até 2x no Cartão: Preço Normal.
(4) 3x Ou Mais no Cartão: 20% de Juros.{'\033[0m'}''')
produto = float(input('Qual o valor do produto? R$ '))
condicao_p = str(input('Qual a condição de pagamento? '))
if '1' in condicao_p:
    print(f'{'\033[1;33m'}O produto no valor de R${produto:.2f} à vista no Dinheiro/Cheque com 10% de desconto {produto - (produto * 10 / 100):.2f}')
elif '2' in condicao_p:
    print(f'{'\033[1;33m'}O produto no valor de R${produto:.2f}, à vista no Cartão com 5% de desconto {produto - (produto * 5 / 100):.2f}')
elif '3' in condicao_p:
    parcela = produto / 2
    print(f'{'\033[1;33m'}O produto no valor de R${produto:.2f} em até 2x no Cartão {parcela:.2f}')
elif '4' in condicao_p:
    total = produto + (produto * 20 / 100)
    total_par = int(input('Quantas parcelas'))
    parcela = total / total_par
    print(F'{'\033[1;33m'}Sua compra será parcelado em {total_par}x de {parcela:.2f}')
    print(f'O produto no valor de {produto:.2f}, com a forma de pagamento 3x ou mais no Cartão com 20% de Juros {produto + (produto * 20 / 100):.2f}')
else:
    print(f'{'\033[1;31m'}OPÇÃO INVÁLIDA de pagamento. Tente novamente!{'\033[0m'}')
```
