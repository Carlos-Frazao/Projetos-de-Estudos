049-Refaça o DESAFIO 009, mostrando a tabuada de um número que o usuário escolher, só que agora utilizando um laço for.

```python
n = int(input('Digite um número, onde será mostrado sua tabuada.'))
print('=-' * 6)
for c in range(1, 11):
    print(f'{'\033[1;32m'}{n} X {c:2} = {n * c:2}{'\033[0m'}')
print('=-' * 6)
```
