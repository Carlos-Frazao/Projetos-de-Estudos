### Fatiamento

```python
# Fatiamento de strings
frase = 'Curso em Vídeo Python'
print(frase)

# Mostra o caractere do índice 9
print(frase[9])

# Mostra os caracteres do índice 9 ao 13, o 14 não é incluso
print(frase[9:14])

# Mostra do índice 9 ao 20, pulando de 2 em 2
print(frase[9:21:2])

# Do início até o índice 4
print(frase[:5])

# Do índice 15 até o final
print(frase[15:])

# Do índice 9 até o final, pulando de 3 em 3
print(frase[9::3])
```

### Análise 

```python
# Analise de strings
frase = 'Curso em Vídeo Python'
print(frase)

# len - Quer dizer comprimento, exibira a quantidade de caracteres na frase.
print(len(frase))

# count - Conta quantas vezes aparece a letra 'o' na frase.
print(frase.count('o'))

# count com fatiamento - Conta quantas vezes aparece a letra 'o' do índice 0 ao 12
print(frase.count('o', 0, 13))

# find - Mostra em que índice começa a palavra 'Vídeo'
print(frase.find('deo'))

# find quando não encontra - Retorna -1 quando não encontra a palavra
print(frase.find('Android'))

# in - Retorna True se a palavra existir na frase
print('Curso' in frase)

```

### Transformação

```python
# Transformação de strings 
frase = 'Curso em Vídeo Python'
print(frase)

# replace - Substitui a palavra 'Python' por 'Android'
print(frase.replace('Python', 'Android'))

# upper - Deixa todas as letras maiúsculas
print(frase.upper())

# lower - Deixa todas as letras minúsculas
print(frase.lower())

# capitalize - Deixa só a primeira letra maiúscula
print(frase.capitalize())

# title - Deixa todas as primeiras letras de cada palavra maiúsculas
print(f'{frase.title()}\n')

segunda_frase = ('   Aprenda Python  ')
print(segunda_frase)

# strip - Remove os espaços em branco do início e do fim da frase
print(segunda_frase.strip())

# rstrip - Remove os espaços em branco do lado direito
print(segunda_frase.rstrip())

# lstrip - Remove os espaços em branco do lado esquerdo
print(segunda_frase.lstrip())
```

### Divisão

```python
# Concatenação de strings
frase = 'Curso em Vídeo Python'

# Onde tiver o espaço, ele adiciona um espaço, ele refaz o indexe, cada palavra vira um índice.
print(frase.split())

# join - Junta as palavras com o que for colocado entre aspas
print('-'.join(frase.split()))
```
