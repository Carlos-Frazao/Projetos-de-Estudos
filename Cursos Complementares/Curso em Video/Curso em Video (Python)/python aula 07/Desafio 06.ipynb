```python
import math # Importando a biblioteca matemática
def main():
    # Entrada de dados
    num = int(input("Informe um número: "))

    # Saída de dados
    print(f"O dobro de {num} vale {num * 2} \nO triplo de {num} vale {num * 3} \nA raiz quadrada de {num} é igual a {math.sqrt(num):.1f}")

if __name__ == "__main__":
    main()
```
