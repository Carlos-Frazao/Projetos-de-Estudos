```python
# Fazer a média com função 
def media(nota1, nota2):
    return (nota1 + nota2) / 2

def main():
    # Entrada de dados
    n1 = float(input("Informe a primmeira nota: "))
    n2 = float(input("Informe a segunda nota: "))
    
    # Saída de dados
    print(f"A média entre {n1} e {n2} é igual a {media(n1,n2):.1f}")

if __name__ == "__main__":
    main()
```
