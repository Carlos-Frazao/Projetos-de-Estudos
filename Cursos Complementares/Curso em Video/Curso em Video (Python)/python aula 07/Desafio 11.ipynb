```python
def main():
    largura = float(input("Largura da parede: "))
    altura = float(input("Altura da parede: "))
    area = largura * altura
    litro = 2 
    # Saída de dados
    print(f"Sua parede tem a dimensão de {largura}X{altura} e ssua área é de {area}m² \nPara pintar essa parede será necessário {area/litro}L de tinta")
    
if __name__ == "__main__":
    main()
```
