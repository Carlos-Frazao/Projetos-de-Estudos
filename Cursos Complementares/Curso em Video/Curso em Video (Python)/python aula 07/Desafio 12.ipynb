```python
def main():
    preco = float(input("Informe o preço do produto: R$"))
    # Saída de dados com o desconto.
    print(f"Novo preço com 5% de desconto R${preco-(preco*5/100):.2f}")
if __name__ == "__main__":
    main()
```

```python
def desconto(preco_produto, porcentagem_desconto):
    return preco_produto - (preco_produto*porcentagem_desconto/100)

def main():
    try:
        preco = float(input("Informe o preço do produto: R$"))
        porcentagem = float(input("Informe a porcentagem de desconto: "))
        # Saída de dados, cálculo com função.
        print(f"Novo preço com {porcentagem}% de desconto R${desconto(preco, porcentagem):.1f}")
    except ValueError:
        print("Por favor, insira um valor numérico válido para o preço e a porcentagem.")
        
if __name__ == "__main__":
    main()
```
